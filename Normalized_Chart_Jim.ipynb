```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
TT = pd.read_csv("https://raw.githubusercontent.com/iconxpie/dataEMBA/6e219c86137bfef5c9def1bc63ea51d537edbf00/Normalized%20Jim", skiprows=0, header=0 )
TT.head()
```

```python
TT.plot(title='Gloves, Pieces, and Shrink Wrap');
```

```python
plt.rcParams['figure.figsize'] = [18, 8]
TT[["Gloves", "Shrink_Wrap"]].plot(title='Pieces: Gloves and Shrink Wrap');
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [18, 8]
TT[["Shrink_Wrap", "Pieces_Delivered"]].plot(title='Shrink Wrap and Pieces Delivered');
plt.show()
```
